### Modeling

### 1) Logistic Regression

#### Train the model with cross validation and create out-of-fold(oof) predictions

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
import numpy as np


X = data.drop(columns=["TARGET"])
y = data["TARGET"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
oof_preds = np.zeros(len(y))
oof_truth = y.values.copy()
```

```python
for fold, (train_idx, val_idx) in enumerate(skf.split(X_scaled, y)):
    X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
    y_train = y.iloc[train_idx]

    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)

    y_val_pred = model.predict_proba(X_val)[:, 1]
    oof_preds[val_idx] = y_val_pred

threshold = 0.5
y_pred_final = (oof_preds >= threshold).astype(int)

auc = roc_auc_score(oof_truth, oof_preds)
precision = precision_score(oof_truth, y_pred_final)
recall = recall_score(oof_truth, y_pred_final)
f1 = f1_score(oof_truth, y_pred_final)

# Print results
print(f"\n--- OOF Evaluation Metrics at Threshold {threshold} ---")
print(f"AUC Score     : {auc:.5f}")
print(f"Precision     : {precision:.5f}")
print(f"Recall        : {recall:.5f}")
print(f"F1 Score      : {f1:.5f}")
```

#### Metrics for different thresholds

```python
thresholds = np.arange(0.05, 0.55, 0.05)
precisions, recalls, f1s = [], [], []

for thresh in thresholds:
    y_pred_thresh = (oof_preds >= thresh).astype(int)
    precisions.append(precision_score(oof_truth, y_pred_thresh))
    recalls.append(recall_score(oof_truth, y_pred_thresh))
    f1s.append(f1_score(oof_truth, y_pred_thresh))

results_df = pd.DataFrame({
    "Threshold": thresholds,
    "Precision": precisions,
    "Recall": recalls,
    "F1 Score": f1s
})

print(results_df)

```

```python
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions, label='Precision', marker='o')
plt.plot(thresholds, recalls, label='Recall', marker='s')
plt.plot(thresholds, f1s, label='F1 Score', marker='^')
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Precision, Recall, and F1 vs Threshold (OOF Predictions) for Logistic Regression")
plt.legend()
plt.grid(True)
plt.xticks(thresholds)
plt.show()
```

#### Best Threshold Metrics

```python
best_idx = np.argmax(f1s)
best_threshold = thresholds[best_idx]

print(f"Best Threshold (max F1): {best_threshold:.2f}")
print(f"Precision: {precisions[best_idx]:.5f}")
print(f"Recall   : {recalls[best_idx]:.5f}")
print(f"F1 Score : {f1s[best_idx]:.5f}")
```

#### Test Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

def preprocess_test_data(app_test,
                         bureau_agg, prev_agg, inst_agg,
                         cc_agg, pos_agg,
                         train_columns):
    test = app_test.copy()

    
    test['AGE_YEARS'] = (-test['DAYS_BIRTH'] / 365).round(1)
    test['EMPLOYED_ANOM'] = test['DAYS_EMPLOYED'] == 365243
    test['DAYS_EMPLOYED_CLEAN'] = test['DAYS_EMPLOYED'].replace(365243, np.nan)
    test['EMPLOYED_YEARS'] = -test['DAYS_EMPLOYED_CLEAN'] / 365
    test.drop(columns='DAYS_EMPLOYED', inplace=True)

   
    test = test.merge(bureau_agg, on='SK_ID_CURR', how='left')
    test = test.merge(prev_agg, on='SK_ID_CURR', how='left')
    test = test.merge(inst_agg, on='SK_ID_CURR', how='left')
    test = test.merge(cc_agg, on='SK_ID_CURR', how='left')
    test = test.merge(pos_agg, on='SK_ID_CURR', how='left')


    test['CREDIT_INCOME_RATIO'] = test['AMT_CREDIT'] / test['AMT_INCOME_TOTAL']
    test['ANNUITY_INCOME_RATIO'] = test['AMT_ANNUITY'] / test['AMT_INCOME_TOTAL']
    test['ANNUITY_CREDIT_RATIO'] = test['AMT_ANNUITY'] / test['AMT_CREDIT']
    test['DAYS_BIRTH'] = -test['DAYS_BIRTH']
    test['DAYS_EMPLOYED_CLEAN'] = -test['DAYS_EMPLOYED_CLEAN']
    test['EMPLOYED_PERCENT'] = test['DAYS_EMPLOYED_CLEAN'] / test['DAYS_BIRTH']

    # Handle missing values
    for col in test.select_dtypes(include=['float64', 'int64']).columns:
        if test[col].isnull().any():
            test[col].fillna(test[col].median(), inplace=True)
    for col in test.select_dtypes(include='object').columns:
        if test[col].isnull().any():
            test[col].fillna(test[col].mode()[0], inplace=True)

    # Categorical encoding
    cat_cols = test.select_dtypes(include='object').columns.tolist()
    le = LabelEncoder()
    for col in cat_cols:
        if test[col].nunique() == 2:
            test[col] = le.fit_transform(test[col])
    
    # One-hot encode low cardinality
    low_card_cat = [col for col in cat_cols if 3 <= test[col].nunique() < 10]
    test = pd.get_dummies(test, columns=low_card_cat, drop_first=True)

    # Encode high-cardinality cols if present
    for col in ['OCCUPATION_TYPE', 'ORGANIZATION_TYPE']:
        if col in test.columns:
            test[col] = le.fit_transform(test[col].astype(str))

    # Reindex to match training columns
    test = test.reindex(columns=train_columns, fill_value=0)

    return test
```
